```python
#scale all values between 0 and 1
from sklearn.preprocessing import MinMaxScaler
scld = MinMaxScaler(feature_range=(0,1))
arr_scld=scld.fit_transform(data)
data_scld = pd.DataFrame(arr_scld,columns=data.columns)
data.head()
data.describe()
data_scld.head()
data_scld.describe()
```

```python
#Create Features and labels
x=data_scld.drop('Outcome',axis=1)
y=data_scld['Outcome']
x.head()
y.head()
```

```python
svc=SVC()
svc.get_params()
```

```python
param_grid= [{'c':[1,10],'Kernel':['rbf'],'gamma':[0.01,1]}]
```

```python
model=GridSearchCV(svc,param_grid,cv=10,scoring='accuracy')
```

```python
model.fit(x,y)
```
