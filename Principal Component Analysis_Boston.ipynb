```python
import numpy as np
import pandas as pd
```

```python
#Import Boston Dataset from UCI site
boston = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data",delim_whitespace=True,header=None)
boston.head()
```

```python
#Add the column names as described in https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.names
boston.columns = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
```

```python
# Find and Print the mean and round it to 2 decimal values
means = np.round(pd.pivot_table(boston,values=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS',
                                               'RAD','TAX','PTRATIO','B','LSTAT','MEDV'],index='MEDV'),2)
means
```

```python
#Drop the label(value to be rpedicted) from X axis
X = boston.drop('MEDV',axis =1)
#
Y = boston['MEDV']
X.head()
Y.head()
```

```python
from sklearn.preprocessing import scale
X_scaled = scale(X)
X_scaled = pd.DataFrame(X_scaled,columns=X.columns)
X_scaled.head()
X_scaled.describe()
```

```python
from sklearn.decomposition import PCA
pca = PCA().fit(X_scaled)
```

```python
str(pca.explained_variance_ratio_)
```

```python
pca = PCA(5).fit(X_scaled)
```

```python
pca.explained_variance_
```

```python
pca_df = pd.DataFrame(pca.components_, columns=X.columns)
pca_df
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
plt.figure(figsize=(10,10))
import seaborn as sns
sns.heatmap(pca_df, cmap='RdYlGn', annot = True)
plt.show()
```

```python
X_scaled_trnsf = pca.fit_transform(X_scaled)
```

```python
X_scaled_trnsf = pd.DataFrame(X_scaled_trnsf, columns=('P1','P2','P3','P4','P5'))
X_scaled_trnsf.head()
X_scaled_trnsf.shape
```
