```python
#To get the count of levels according to AQI
btm_good = btm.loc[btm['BNOx']<=40]
btm_sat = btm.loc[(btm['BNOx']>40) & (btm['BNOx']<=80)]
btm_mod = btm.loc[(btm['BNOx']>80) & (btm['BNOx']<=180)]
btm_poor = btm.loc[(btm['BNOx']>180) & (btm['BNOx']<=280)]
btm_verypoo = btm.loc[(btm['BNOx']>280) & (btm['BNOx']<=400)]
btm_severe = btm.loc[btm['BNOx']>400]
```

```python
AQI_BTM = [['Good', len(btm_good)], ['Satisfactory', len(btm_sat)], ['Moderate', len(btm_mod)],['Poor', len(btm_poor)],['Verypoor', len(btm_verypoo)],['Severe', len(btm_severe)]] 
AQI_BTM = pd.DataFrame(AQI_BTM, columns = ['Level', 'BCount']) 
AQI_BTM['Percentage'] =  AQI_BTM['BCount']/AQI_BTM['BCount'].sum()
AQI_BTM= AQI_BTM.rename(columns = {'Percentage':'BPercentage'})
AQI_BTM.head(6)
```

```python
p_good = peenya.loc[peenya['PNOx']<=40]
p_sat =  peenya.loc[( peenya['PNOx']>40) & ( peenya['PNOx']<=80)]
p_mod =  peenya.loc[( peenya['PNOx']>80) & ( peenya['PNOx']<=180)]
p_poor = peenya.loc[(peenya['PNOx']>180) & (peenya['PNOx']<=280)]
p_verypoo = peenya.loc[(peenya['PNOx']>280) & (peenya['PNOx']<=400)]
p_severe = peenya.loc[peenya['PNOx']>400]
```

```python
len(p_sat)
```

```python
AQI_Peenya = [['Good', len(p_good)], ['Satisfactory', len(p_sat)], ['Moderate', len(p_mod)],['Poor', len(p_poor)],['Verypoor', len(p_verypoo)],['Severe', len(p_severe)]] 
AQI_Peenya = pd.DataFrame(AQI_Peenya, columns = ['Level', 'PCount']) 
AQI_Peenya['PPercentage'] =  AQI_Peenya['PCount']/AQI_Peenya['PCount'].sum()
AQI_Peenya.head(6)
```

```python
k_good = kb.loc[kb['KNOx']<=40]
k_sat = kb.loc[(kb['KNOx']>40) & (kb['KNOx']<=80)]
k_mod = kb.loc[(kb['KNOx']>80) & (kb['KNOx']<=180)]
k_poor = kb.loc[(kb['KNOx']>180) & (kb['KNOx']<=280)]
k_verypoo = kb.loc[(kb['KNOx']>280) & (kb['KNOx']<=400)]
k_severe = kb.loc[kb['KNOx']>400]
```

```python
AQI_Kb = [['Good', len(k_good)], ['Satisfactory', len(k_sat)], ['Moderate', len(k_mod)],['Poor', len(k_poor)],['Verypoor', len(k_verypoo)],['Severe', len(k_severe)]] 
AQI_Kb = pd.DataFrame(AQI_Kb, columns = ['Level', 'KCount']) 
AQI_Kb['KPercentage'] =  AQI_Kb['KCount']/AQI_Kb['KCount'].sum()
AQI_Kb.head(6)
```

```python
AQI = pd.concat([AQI_BTM ,AQI_Peenya['PPercentage'],AQI_Kb['KPercentage']], axis = 1)
AQI.head()
```

```python
AQI = AQI.drop(['BCount'], axis = 1)
AQI.head()
```

```python
AQI = AQI.set_index("Level")
AQI
```

```python
AQI= AQI.rename(columns = {'BPercentage':'BTM Layout','PPercentage':'Peenya',"Kpercentage":'Kadubeesanahalli'})
```

```python
AQI.iplot(kind='bar', xTitle='Levels', yTitle='Percentage of count of days',
    title= 'AQI Index')
```
