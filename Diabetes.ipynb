```python
#Create new variables
df['glRBMI'] = df['Glucose']/df['BMI']
df['BMIRage'] = df['BMI']/df['Age']
df['glRage']=df['Glucose']/df['Age']
```

```python
df.head()
```

```python
#Bin variables Glucose
import numpy as np
df['glBin'] = np.where(df['Glucose'] <=125,'<=125', (np.where(df['Glucose']> 150 ,'>150', '125-150')))
```

```python
df.head()
```

```python
pd.pivot_table(df, values="Glucose", index=["glBin"], columns="Outcome", aggfunc = "count",fill_value=0)
```

```python
#Bin variables BMI
import numpy as np
df['bmiBin'] = np.where(df['BMI'] <=25,'<=25', (np.where(df['BMI']> 35 ,'>35', '25-35')))
```

```python
df.head()
```

```python
pd.pivot_table(df, values="BMI", index=["bmiBin"], columns="Outcome", aggfunc = "count",fill_value=0)
```

```python
#Bin variables Age
import numpy as np
df['ageBin'] = np.where(df['Age'] <=30,'<=30', (np.where(df['Age']> 45 ,'>45', '30-45')))
```

```python
df.head()
```

```python
pd.pivot_table(df, values="Age", index=["ageBin"], columns="Outcome", aggfunc = "count",fill_value=0)
```
